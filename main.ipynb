## Maksimalpunkt

Vi ser på funksjonen
f(x) = e^(-x/4) * arctan(x)

### Derivasjon

Jeg brukte produktregelen:

f'(x) = e^(-x/4) * 1/(1 + x^2) - (1/4)*e^(-x/4)*arctan(x)

Jeg faktoriserte ut e^(-x/4):

f'(x) = e^(-x/4) * (1/(1 + x^2) - (1/4)*arctan(x))

Setter f'(x) = 0:

Siden e^(-x/4) aldri er 0, får vi:

arctan(x) - 4/(1 + x^2) = 0

---

### Numerisk løsning

Jeg løste likningen med halveringsmetoden.

Først testet jeg noen verdier:
g(1) < 0
g(2) > 0

Dermed ligger løsningen mellom 1 og 2.

Ved å halvere intervallet flere ganger fikk jeg:

x ≈ 1.3917

---

### Toppunkt

Setter inn i funksjonen:

f(1.3917) ≈ 0.5290

Toppunktet er:

(1.3917, 0.5290)

---

### Plot

Jeg plottet funksjonen i Python og markerte toppunktet for å sjekke svaret.


```python
import numpy as np

def g(x):
    return np.arctan(x) - 4/(x**2 + 1)

a = 1
b = 2

tol = 1e-5

while (b - a) > tol:
    m = (a + b) / 2
    
    if g(a) * g(m) < 0:
        b = m
    else:
        a = m

x_max = (a + b) / 2

def f(x):
    return np.exp(-x/4) * np.arctan(x)

y_max = f(x_max)

print(f"x ≈ {x_max:.4f}")
print(f"f(x) ≈ {y_max:.4f}")
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)
y = f(x)

plt.plot(x, y)
plt.scatter(x_max, y_max)

plt.grid()
plt.show()
```
