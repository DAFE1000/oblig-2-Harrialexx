```python
import numpy as np

def g(x):
    return np.arctan(x) - 4/(x**2 + 1)

a = 1
b = 2

tol = 1e-5  

while (b - a) > tol:
    m = (a + b) / 2
    
    if g(a) * g(m) < 0:
        b = m
    else:
        a = m

x_topp = (a + b) / 2
print("x =", round(x_topp, 4))
```

```python
f = lambda x: np.exp(-x/4) * np.arctan(x)

y_topp = f(x_topp)
print("f(x) =", round(y_topp, 4))
```
